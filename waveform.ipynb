```python

import matplotlib.pyplot as plt

# Load data from the Xyce output file
data = []

with open("work/vt_0.65_25.0_delay1_INVx1_A01_Y10_0.000675_0.005992_energy.sp.prn", "r") as f:
    # Skip the header line
    header = f.readline()
    
    # Read the rest of the lines as data
    for line in f:
        # Split line into elements and convert to floats
        elements = line.split()
        try:
            # Convert all elements to float except the first column which is index
            data.append([float(el) for el in elements[1:]])
        except ValueError:
            continue  # Skip any line that can't be converted

# Transpose the data to plot
data = list(zip(*data))

# Plot the waveform
plt.figure(figsize=(10,6))
plt.plot(data[0], data[1], label='V(VIN)')
plt.title("Circuit Response")
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.grid(True)
plt.legend()
plt.show()

# plt.figure(figsize=(10,6))
# plt.plot(data[0], data[3], label='V(VCIN)')
# plt.title("Circuit Response")
# plt.xlabel("Time (s)")
# plt.ylabel("Voltage (V)")
# plt.grid(True)
# plt.legend()
# plt.show()

plt.figure(figsize=(10,6))
plt.plot(data[0], data[2], label='V(VOUT)')
plt.title("Circuit Response")
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.grid(True)
plt.legend()
plt.show()

if(len(data)>5):
    plt.figure(figsize=(10,6))
    plt.plot(data[0], data[3], label='I(VDD_DYN)')
    plt.title("Circuit Response")
    plt.xlabel("Time (s)")
    plt.ylabel("Voltage (V)")
    plt.grid(True)
    plt.legend()
    plt.show()

    plt.figure(figsize=(10,6))
    plt.plot(data[0], data[4], label='I(VSS_DYN)')
    plt.title("Circuit Response")
    plt.xlabel("Time (s)")
    plt.ylabel("Voltage (V)")
    plt.grid(True)
    plt.legend()
    plt.show()

    plt.figure(figsize=(10,6))
    plt.plot(data[0], data[5], label='I(VIN)')
    plt.title("Circuit Response")
    plt.xlabel("Time (s)")
    plt.ylabel("Voltage (V)")
    plt.grid(True)
    plt.legend()
    plt.show()
```
